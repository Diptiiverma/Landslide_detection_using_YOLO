```python
#Python auto split data for test

import os
import random
import shutil

# paths
train_images = "/content/KTproject-11/train/images"
train_labels = "/content/KTproject-11/train/labels"

test_images = "/content/KTproject-11/test/images"
test_labels = "/content/KTproject-11/test/labels"

os.makedirs(test_images, exist_ok=True)
os.makedirs(test_labels, exist_ok=True)

# percentage for test split
test_ratio = 0.15   # 15%

image_files = os.listdir(train_images)
random.shuffle(image_files)

test_count = int(len(image_files) * test_ratio)
test_files = image_files[:test_count]

for img in test_files:
    label = img.rsplit(".", 1)[0] + ".txt"

    shutil.move(os.path.join(train_images, img), os.path.join(test_images, img))
    shutil.move(os.path.join(train_labels, label), os.path.join(test_labels, label))

print("Test split created successfully!")
```
